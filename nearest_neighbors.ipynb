```python
from scipy import ndimage
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import numpy as np
%matplotlib inline
```

```python
A = ndimage.imread('test.jpg')
A = A / 255.
original_shape = A.shape
```

```python
plt.imshow(A)
```

```python
A = np.reshape(A, (original_shape[0] * original_shape[1], original_shape[2]))
```

```python
def find_closest_centroids(X, centroids):  
    m = X.shape[0]
    k = centroids.shape[0]
    idx = np.zeros(m)

    for i in range(m):
        min_dist = 1000000
        for j in range(k):
            dist = np.sum((X[i,:] - centroids[j,:]) ** 2)
            if dist < min_dist:
                min_dist = dist
                idx[i] = j

    return idx

def compute_centroids(X, idx, k):  
    m, n = X.shape
    centroids = np.zeros((k, n))

    for i in range(k):
        indices = np.where(idx == i)
        centroids[i,:] = (np.sum(X[indices,:], axis=1) / len(indices[0])).ravel()

    return centroids

def run_k_means(X, num_centroids, max_iters):  
    initial_centroids = X[np.random.choice(len(X),size=num_centroids,replace=False)]
    m, n = X.shape
    k = initial_centroids.shape[0]
    idx = np.zeros(m)
    centroids = initial_centroids

    for i in range(max_iters):
        idx = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, k)

    return idx, centroids
```

```python
fig = plt.figure(figsize=(50,50))

for i in range(1,10):
    fig.add_subplot(3,3,i)
    idx, centroids = run_k_means(A, i+1, 10)
    A_new = np.array([centroids[int(i)] for i in idx])
    A_new = np.reshape(A_new, original_shape)
    plt.imshow(A_new)
                     
```

```python
fig = plt.figure(figsize=(50,50))

for idx, centroid in enumerate(centroids[:-1]):
    fig.add_subplot(3,3,idx+1)
    A_new_new = A_new.copy()
    A_new_new = A_new_new.reshape(original_shape[0] * original_shape[1], original_shape[2])
    A_new_new[A_new_new != centroid] = 0
    A_new_new[A_new_new == centroid] = 1

    A_new_new = A_new_new.reshape(original_shape)
    plt.imshow(A_new_new)
```
